```python
import matplotlib.pyplot as plt
import numpy as np
import re, os, pickle
from collections import defaultdict
```

```python
list_of_models = [item for item in os.listdir('models_v2/') if 'pkl' in item]
```

```python
metrics = ['val_loss', 'val_mcc_k', 'loss', 'mcc_k']
projection_dict = {}


for model_ in list_of_models:
    with open('models_v2/' + model_, 'rb') as fp:
        model_data = pickle.load(fp)
        
        metric_dict = {'val_loss':[], 'val_mcc_k':[], 'loss':[], 'mcc_k':[]}

        for metric in metrics:
            for k,v in model_data['training'].items():
                metric_dict[metric] += v['results'][metric]
        
        projection_dict.setdefault(model_data['feature_config']['time_steps'], []).append(metric_dict)

fig = plt.figure(figsize=(15, 10))
for st_ in enumerate(metrics):   
    plt.subplot(2, 2, st_[0] + 1)
    qu1 = []
    qu2 = []
    for key in projection_dict.keys():
        for item in projection_dict[key]:
            if len(item[st_[1]]) == 100:
                qu1.append(item[st_[1]])
            else:
                qu2.append(item[st_[1]])
        thing1 = np.mean(np.squeeze(np.array(qu1)), axis = 0)
        thing2 = np.mean(np.squeeze(np.array(qu2)), axis = 0)
        plt.plot(thing1, label = key, linewidth = 3)
        plt.plot(thing2, label = key, linewidth = 1)
    plt.title(st_[1])
    plt.legend()
```

```python
metrics = ['val_loss', 'val_mcc_k', 'loss', 'mcc_k']
projection_dict = {}


for model_ in list_of_models:
    with open('models_v2/' + model_, 'rb') as fp:
        model_data = pickle.load(fp)
        
        metric_dict = {'val_loss':[], 'val_mcc_k':[], 'loss':[], 'mcc_k':[]}

        for metric in metrics:
            for k,v in model_data['training'].items():
                metric_dict[metric] += v['results'][metric]
        
    projection_dict[re.findall("[a-zA-Z0-9]*", model_)[0]] = {'metrics': metric_dict, 
                                                              'model_config': model_data['model_config'],
                                                              'train_config': model_data['train_config'],
                                                              'feature_config': model_data['feature_config']
                                                             }
    
```

```python
X = []
X1 = []
for k,v in projection_dict.items():
    X.append((sum(v['metrics']['loss']), sum(v['metrics']['val_mcc_k'])))
    X1.append((k, sum(v['metrics']['val_mcc_k'])*sum(v['metrics']['loss'])))
```

```python
plt.scatter(*zip(*X))
plt.xlabel('loss')
plt.ylabel('val_mcc_k')
plt.show()
```

```python
bb = sorted(X1, key=lambda x:-x[1])[0:10]
aa = [item[0] for item in bb]
```
